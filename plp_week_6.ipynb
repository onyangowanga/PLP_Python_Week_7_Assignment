# Task 1: Load and Explore the Dataset

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load a Dataset 

```python
try:
    columns = ['sepal.length','sepal.width','petal.length','petal.width','variety']
    
    # Load the dataset
    file_path = 'iris.csv' 
    data = pd.read_csv(file_path, header=0, names=columns)
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found. Please check the file path and try again.")
except pd.errors.ParserError:
    print(f"Error: The file '{file_path}' could not be parsed. Please ensure it is a valid CSV file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

### Display the first 5 rows

```python
print("First 5 rows of the dataset:")
print(data.head())
```

## Inspect the Dataset

### Check the data types

```python
print("\nDataset Information:")
print(data.info())
```

### Check for missing values

```python
print("\nMissing Values in Each Column:")
print(data.isnull().sum())
```

### Drop missing Rows

```python
data.dropna(inplace=True)
```

### Check to ensure no more missing values

```python
print("\nDataset after cleaning:")
print(data.head())
```

# Task 2: Basic Data Analysis

#### Basic statistics for numerical columns

```python
print("Basic Statistics:")
print(data.describe())
```

#### Group by 'variety' and compute mean for each group

```python
grouped_means = data.groupby('Species').mean()

print("\nMean values grouped by species:")
print(grouped_means)
```

# Key Observations
### Sepal vs. Petal Sizes:

The sepals (the outer part of the flower) are generally larger than the petals in both length and width. 
For the "Setosa" variety, the petals are noticeably smaller and quite uniform in size.

#### Consistency in Measurements:

The measurements for "Setosa" seem very consistent, especially for petal size, as there's very little variation. This could mean that "Setosa" flowers tend to have a predictable and uniform structure.

#### Unique Characteristics of "Setosa":

Compared to other varieties of the Iris plant, "Setosa" is likely to stand out due to its smaller petal size and more compact proportions.

## Interesting Patterns
The petal dimensions of "Setosa" are much smaller and don’t vary much between plants. This interesting trait might make it easier to classify this variety compared to others.
The sepal dimensions show slightly more variation but still seem consistent enough to suggest a clear pattern for "Setosa."

# What This Means
These patterns suggest that "Setosa" is a distinct and predictable variety of the Iris plant, making it easier to identify. 
If we were to include other varieties in the analysis, comparing their measurements might reveal even more distinct characteristics.



# Task 3: Data Visualization

## 1. Line Chart

#### Line chart for sepal length

```python
plt.figure(figsize=(8, 5))
plt.plot(data.index, data['sepal.length'], marker='o', linestyle='-', color='b', label='Sepal Length')
plt.title('Sepal Length Trend Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length')
plt.legend()
plt.grid()
plt.show()
```

## 2. Bar Chart

#### Bar chart for average sepal length per species

```python
# Calculate the average sepal length for each variety
avg_sepal_length = data.groupby('variety')['sepal.length'].mean()

# Plot the results
plt.figure(figsize=(8, 5))
avg_sepal_length.plot(kind='bar', color='orange', alpha=0.8)
plt.title('Average Sepal Length per Variety')
plt.xlabel('Variety')
plt.ylabel('Average Sepal Length')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## 3. Histogram

#### Histogram for sepal length

```python
plt.figure(figsize=(8, 5))
plt.hist(data['sepal.length'], bins=8, color='purple', alpha=0.7, edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

## 4. Scatter Plot

#### Scatter plot for sepal length vs petal length

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['sepal.length'], data['petal.length'], color='green', alpha=0.6, label='Data Points')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.grid()
plt.show()
```
